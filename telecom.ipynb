```python
telecom['BEGIN'] = telecom['BEGIN'].str.replace("/",", ")
telecom['END'] = telecom['END'].str.replace("/",", ")
def jours(debut,fin):
    import datetime
    born = datetime.date(debut)
    today = datetime.date(fin)
    diff = today - born
    print(diff.total_seconds())
jours(telecom['BEGIN'], telecom['END'])
```

```python
def difference_dates(debut, fin):
    from datetime import datetime
    x = datetime.now(debut)
    y = datetime.now(fin)
    delta = y - x
    return delta 
telecom['daysPeriod'] = telecom['BEGIN'],telecom['END'].apply(difference_dates)
telecom
```
